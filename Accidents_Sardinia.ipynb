```python
import requests
import json

# URL of the Overpass API
overpass_url = "https://overpass-api.de/api/interpreter"

# Overpass Query to retrieve hospitals, schools, and police stations in Sardinia
overpass_query = """
[out:json][timeout:60];
area["name"="Sardegna"]->.searchArea;
(
  node["amenity"="hospital"](area.searchArea);
  node["amenity"="school"](area.searchArea);
  node["amenity"="police"](area.searchArea);
);
out body;
"""

# Send a request to the API
response = requests.get(overpass_url, params={'data': overpass_query})

# Check if the request was successful
if response.status_code == 200:
    data = response.json()  # Convert the response to JSON format
    
    # Save the JSON response to a file
    with open("infrastrutture_sardegna.json", "w", encoding="utf-8") as file:
        json.dump(data, file, indent=4, ensure_ascii=False)

    print("Data successfully saved in 'infrastrutture_sardegna.json'")
else:
    print("Error in the request:", response.status_code)
```

```python
# Load the JSON file
with open("infrastrutture_sardegna.json", "r", encoding="utf-8") as file:
    data = json.load(file)
```

```python
# Extract elements and transform them into a DataFrame
rows = []
for element in data.get("elements", []):
    if element["type"] == "node":  # Filter only nodes (ignore ways and relations)
        row = {
            "ID": element["id"],  # Unique identifier of the infrastructure
            "Latitudine": element["lat"],  # Latitude coordinate
            "Longitudine": element["lon"],  # Longitude coordinate
            "Nome": element["tags"].get("name", "N/A"),  # Name of the infrastructure
            "Tipo": element["tags"].get("amenity", "N/A"),  # Type of infrastructure (hospital, school, police station)
            "Indirizzo": element["tags"].get("addr:street", "N/A"),  # Street address (if available)
            "Numero Civico": element["tags"].get("addr:housenumber", "N/A"),  # House number (if available)
            "Telefono": element["tags"].get("phone", "N/A")  # Contact phone number (if available)
        }
        rows.append(row)

# Create a DataFrame from the extracted data
df_infrastrutture = pd.DataFrame(rows)
print(df_infrastrutture)
```

```python
# Count missing values (including "N/A" and empty strings)
df_infrastrutture_replaced = df_infrastrutture.replace(["", "N/A"], pd.NA)
missing_values = df_infrastrutture_replaced.isnull().sum()

print("Missing values for each column:\n", missing_values) 
```

<span style="font-family: 'Times New Roman'; font-size: 20px; font-weight: bold;">Data Integration</span><br><span style="font-family: 'Times New Roman'; font-size: 18px;">Assigning Provinces to Infrastructure in Sardinia</span><br>
<span style="font-family: 'Times New Roman'; font-size: 16px;"> This script uses **reverse geocoding** with the **Nominatim** service from **OpenStreetMap** to assign the correct **province** to each infrastructure (hospitals, schools, police stations) based on latitude and longitude. </span>



```python
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut
import pandas as pd
import time

# Initialize the geocoder
geolocator = Nominatim(user_agent="incidenti_geocoding", timeout=10)

# Function to retrieve the province based on latitude and longitude
def get_provincia(lat, lon, retries=3):
    """Retrieves the correct province from Nominatim, handling errors and timeouts."""
    try:
        # Perform reverse geocoding to get location details
        location = geolocator.reverse((lat, lon), exactly_one=True, language="it")
        if location and "address" in location.raw:
            return location.raw["address"].get("county")  # Extract the province
    except GeocoderTimedOut:
        if retries > 0:
            time.sleep(10)  # Wait 10 seconds before retrying
            return get_provincia(lat, lon, retries - 1)
        else:
            print(f"Timeout for {lat}, {lon}. No retries left.")
            return None
    except Exception as e:
        print(f"Error for {lat}, {lon}: {e}")
        return None

# Apply the function to the DataFrame to assign provinces
df_infrastrutture["provincia"] = df_infrastrutture.apply(
    lambda row: get_provincia(row["Latitudine"], row["Longitudine"]), axis=1
)

print(df_infrastrutture[["Latitudine", "Longitudine", "provincia"]])
```
